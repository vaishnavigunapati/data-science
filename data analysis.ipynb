```python
# custampize the line plot 
x=np.arange(1,5,0.1)
y=2*x+5
plt.plot(x,y,color='r')
plt.title('line axis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
 
```

```python
# subdivision of each tble
x=np.arange(1,5,0.1)
y1=2*x+5
y2=3*x+10
 
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
# bar plot
fruit={'apple':30,'mango':40,'orange':50}
names=list(fruit.keys())
quantity=list(fruit.values())
```

```python
names,quantity
```

```python
plt.bar(names,quantity)
plt.title('fruit shop')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#  scatter -plot
x=[20,30,40,50,60,70]
a=[1,2,3,4,5,6] 
b=[4,5,6,7,8,9]
plt.figure(figsize=(10,5))
plt.scatter(x,a,s=400,marker='3')
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('x vs ab')
plt.xlabel('x')
plt.ylabel('y')
plt.show

    
```
